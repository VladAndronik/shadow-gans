```
%matplotlib inline
import matplotlib.pyplot as plt
```

```
import glob
import numpy as np
from tqdm import tqdm_notebook
```

```
root_dir = 'data/ISTD_Dataset/'
```

```
os.makedirs(os.path.join(root_dir, 'val', 'val_A'))
os.makedirs(os.path.join(root_dir, 'val', 'val_C'))
```

```
N_VAL = 100
train_A_files = glob.glob(os.path.join(root_dir, 'train', 'train_A', '*.png'))
train_C_files = glob.glob(os.path.join(root_dir, 'train', 'train_C', '*.png'))
```

```
np.random.seed(42)
ind_val_files = np.random.choice(np.arange(0, len(train_A_files)), size=N_VAL, replace=False)
assert len(np.unique(ind_val_files)) == N_VAL
```

```
val_A_files = np.array(train_A_files)[ind_val_files].tolist()
val_C_files = []
for path in val_A_files:
  val_C_files.append(path.replace('train_A', 'train_C'))
```

```
plt.subplots(figsize=(20,10))
counter = 1
for i in tqdm_notebook(range(0, len(val_A_files), 2), total=len(val_A_files) // 2):
  img = plt.imread(val_A_files[i])
  plt.subplot(len(val_A_files) // 2, 5, counter)
  plt.imshow(img)
  counter += 1

plt.subplots_adjust(top=10);
```

```
from shutil import copyfile
```

```
_ = [os.remove(path) for path in glob.glob(os.path.join(root_dir, 'val', 'val_A', '*.png'))]
_ = [os.remove(path) for path in glob.glob(os.path.join(root_dir, 'val', 'val_C', '*.png'))]
```

```
assert len(os.listdir(os.path.join(root_dir, 'val', 'val_A'))) == 0
assert len(os.listdir(os.path.join(root_dir, 'val', 'val_C'))) == 0
```

```
for path_A, path_C in tqdm_notebook(zip(val_A_files, val_C_files), total=N_VAL):
  val_A_dir = os.path.join(root_dir, 'val', 'val_A', os.path.basename(path_A))
  val_C_dir = os.path.join(root_dir, 'val', 'val_C', os.path.basename(path_C))
  copyfile(path_A, val_A_dir)
  copyfile(path_C, val_C_dir)
```

```
assert len(os.listdir(os.path.join(root_dir, 'val', 'val_A'))) == N_VAL
assert len(os.listdir(os.path.join(root_dir, 'val', 'val_C'))) == N_VAL
```

```
_ = [os.remove(path) for path in val_A_files]
_ = [os.remove(path) for path in val_C_files]
```

```
assert len(os.listdir(os.path.join(root_dir, 'train', 'train_A'))) == len(train_A_files) - N_VAL
assert len(os.listdir(os.path.join(root_dir, 'train', 'train_C'))) == len(train_A_files) - N_VAL
```
